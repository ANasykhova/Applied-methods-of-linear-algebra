```python
import numpy as np

P = np.array([[0, -2 / 19, -7/ 19],
             [-1 / 5, 0, -9 / 25],
             [-4 / 22, -3 / 22, 0]])
b = np.array([2,9,5])

x = np.array([0, 0, 0])
i = 1
while np.any(np.abs(P @ x + b - x) > 0.01):
    print(i)
    i += 1
    x = P @ x + b
    print(x)
print()
print(x)
```

```python
A = np.array([[1,1,0,1,0],
              [1,0,1,0,0],
              [0,1,1,0,0],
              [1,1,1,1,0],
              [1,0,0,1,1]])

sum = np.sum(A, axis=0)
P = A / sum
print(P)

beta = 0.15
Q = (1 - beta) * P + beta / P.shape[0] * np.ones(P.shape)
x = np.ones(P.shape[0]) / P.shape[0]
for i in range(20):
    print(x)
    x = Q @ x
```
